```python
df = pd.DataFrame(data)

print(df.columns)
print(df.info())
print(df.isnull().sum())
df = df.drop(columns=['postcode','full_postcode','date_added'])
```

```python

#encoding
le = LabelEncoder()

# Fit and transform the 'type' column
df['type_encoded'] = le.fit_transform(df['propertySubType'])
#df['branch_encoded'] = le.fit_transform(df['branchDisplayName'])
print(df[['propertySubType', 'type_encoded']])

df = pd.get_dummies(df, columns=['New Home'])
df = df.rename(columns={'New Home_0.0': 'New Home_0', 'New Home_1.0': 'New Home_1'})
```

```python
from transformers import TapasForQuestionAnswering, TapasTokenizer
import torch

# Load the TAPAS model and tokenizer
model_name = "google/tapas-base"
model = TapasForQuestionAnswering.from_pretrained(model_name)
tokenizer = TapasTokenizer.from_pretrained(model_name)

# Prepare the input data
# Prepare the input data
def prepare_input(df):
    # Create the input dictionary
    input_dict = {
        "table": df,
        "queries": ["What is the predicted price?"] * len(df)
    }
    
    return input_dict

# Tokenize the input data
def tokenize_input(input_dict, batch_size=100):
    inputs = []
    for i in range(0, len(input_dict["table"]), batch_size):
        batch_dict = {
            "table": input_dict["table"].iloc[i:i+batch_size],
            "queries": input_dict["queries"][i:i+batch_size]
        }
        batch_inputs = tokenizer(
            table=batch_dict["table"],
            queries=batch_dict["queries"],
            padding="max_length",
            return_tensors="pt"
        )
        inputs.append(batch_inputs)
    
    return inputs

# Make predictions
def predict(model, inputs):
    predicted_prices = []
    for batch_inputs in inputs:
        with torch.no_grad():
            outputs = model(**batch_inputs)
            batch_predicted_prices = outputs.logits.argmax(-1).tolist()
            predicted_prices.extend(batch_predicted_prices)
    
    return predicted_prices

# Prepare the input data for prediction
input_dict = prepare_input(df_x)

# Tokenize the input data
inputs = tokenize_input(input_dict)

# Make predictions
predicted_prices = predict(model, inputs)

# Add the predicted prices to the DataFrame
df["predicted_price"] = predicted_prices

# Evaluate the model
from sklearn.metrics import mean_absolute_error, mean_squared_error

mae = mean_absolute_error(df_y, df["predicted_price"])
mse = mean_squared_error(df_y, df["predicted_price"])
rmse = mse ** 0.5

print(f"Mean Absolute Error: {mae}")
print(f"Mean Squared Error: {mse}")
print(f"Root Mean Squared Error: {rmse}")
```

```python
import numpy as np
import torch
from pytorch_tabnet.tab_model import TabNetRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score

# Scale the features
X_scaler = StandardScaler()
X_train_scaled = X_scaler.fit_transform(X_train)
X_test_scaled = X_scaler.transform(X_test)

# Scale the target variable
y_scaler = StandardScaler()
y_train_scaled = y_scaler.fit_transform(y_train.values.reshape(-1, 1))
y_test_scaled = y_scaler.transform(y_test.values.reshape(-1, 1))

# Define and train the TabNet model
tabnet_model = TabNetRegressor(
    n_d=64, n_a=64, n_steps=5,
    gamma=1.5, n_independent=2, n_shared=2,
    lambda_sparse=1e-3, 
    optimizer_fn=torch.optim.Adam,
    optimizer_params=dict(lr=2e-2),
    mask_type='entmax',
    scheduler_params=dict(mode="min",
                          patience=5,
                          min_lr=1e-5,
                          factor=0.9),
    scheduler_fn=torch.optim.lr_scheduler.ReduceLROnPlateau,
    verbose=10
)

tabnet_model.fit(
    X_train=X_train_scaled, y_train=y_train_scaled,
    eval_set=[(X_test_scaled, y_test_scaled)],
    max_epochs=100, patience=10,
    batch_size=1024, virtual_batch_size=128
)

# Make predictions
y_pred_scaled = tabnet_model.predict(X_test_scaled)

# Inverse transform predictions to original scale
y_pred = y_scaler.inverse_transform(y_pred_scaled)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R-squared Score: {r2}")

# Print some predicted vs actual prices
print("\nPredicted vs Actual Prices:")
for pred, actual in zip(y_pred[:20], y_test.values[:20]):
    print(f"Predicted: {pred[0]:.2f}, Actual: {actual:.2f}")

# Calculate and print mean absolute error
mae = np.mean(np.abs(y_pred - y_test.values.reshape(-1, 1)))
print(f"\nMean Absolute Error: {mae:.2f}")

# Calculate and print mean percentage error
mpe = np.mean(np.abs((y_test.values.reshape(-1, 1) - y_pred) / y_test.values.reshape(-1, 1))) * 100
print(f"Mean Percentage Error: {mpe:.2f}%")

# Feature importance
feature_importances = tabnet_model.feature_importances_
for feature, importance in zip(df_x.columns, feature_importances):
    print(f"{feature}: {importance}")
```
